```python
# Selecting the subset of columns to test the CNN 

subset = full_tracks.filter(regex='chroma_cens_kurtosis|yt_views', axis=1)

display(subset.head())

display(len(subset.columns))
```

```python
X1 = subset.drop(columns='yt_views')
Y1 = subset['yt_views']

y_dim = len(X1.columns)

# Split datasets to multiple parts for training, validation and test purposes. 
nb_samples = len(subset)
valid_split = 0.2
test_split = 0.2
X1_train = X1[0:int(nb_samples*(1-valid_split-test_split))]
Y1_train = Y1[0:int(nb_samples*(1-valid_split-test_split))]
X1_valid = X1[int(nb_samples*(1-valid_split-test_split)):int(nb_samples*(1-test_split))]
Y1_valid = Y1[int(nb_samples*(1-valid_split-test_split)):int(nb_samples*(1-test_split))]
X1_test  = X1[int(nb_samples*(1-test_split)):]
Y1_test  = Y1[int(nb_samples*(1-test_split)):]
```

```python
from sklearn.preprocessing import StandardScaler

# Use standard scaler to standardize features by removing the mean and scaling to unit variance.

scaler = StandardScaler().fit(X1_train)

X1_train = scaler.transform(X1_train)
X1_valid = scaler.transform(X1_valid)
X1_test  = scaler.transform(X1_test)

# Needed to reshape the 1D arrays to 2D arrays.

Y1_train = Y1_train.values.reshape(-1, 1)
Y1_valid = Y1_valid.values.reshape(-1, 1)
Y1_test = Y1_test.values.reshape(-1, 1)

scaler = StandardScaler().fit(Y1_train)

Y1_train = scaler.transform(Y1_train)
Y1_valid = scaler.transform(Y1_valid)
Y1_test  = scaler.transform(Y1_test)
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.optimizers import SGD, Adam

# Set random seeds.
np.random.seed(1)
from tensorflow import set_random_seed
set_random_seed(2)

from keras import regularizers

col_num = len(subset.columns) - 1
```

```python
from keras.models import Sequential
from keras.layers.core import Activation, Dense, Flatten
from keras.optimizers import SGD
import matplotlib.pyplot as plt
from keras.layers.convolutional import  Conv1D, MaxPooling1D
```

```python
# Create the CNN here
model = Sequential()
model.add(Conv1D(filters=32, kernel_size=7, input_shape=(y_dim, 1)))
model.add(MaxPooling1D(pool_size=3 ))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mse', optimizer='adam', metrics=['mse'])
model.summary()
```

```python
# Need to resize the input data because the input shape of Keres Conv1D expects 3d
# (number of the data, lenght of the data, channels of the data)
X1_train = X1_train.reshape(24391, y_dim, 1)
X1_valid = X1_valid.reshape(8130, y_dim, 1)
```

```python
model.fit(X1_train, Y1_train, epochs=10, batch_size=12, validation_data=(X1_valid, Y1_valid), verbose=2)
```
