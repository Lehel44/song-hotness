```python
# This csv contains the music genre for the most songs
genres = pd.read_csv('genres.csv')
genres
```

```python
# Makes categorical variable of the ten biggest genre.
new_cols = genres.nlargest(10, '#tracks')['title'].values
main_genres_ids = genres.nlargest(10, '#tracks')['genre_id'].values
new_cols
```

```python
# Add new colls to the tracks dataframe for the categorical variables
genre_cols = [];
for i in new_cols:
    tracks['genre_'+i] = pd.Series(np.zeros(106574, dtype=np.int))
    genre_cols.append('genre_'+i)
```

```python
tracks.head()
```

```python
# delete the fisrt and last character of these strings: the '[' and ']'
tracks['track_genres_all'] = tracks['track_genres_all'].apply(lambda x: x[1:-1])
# than split the string, the separator is: ', ' (comma plus space)
tracks['track_genres_all'] = tracks['track_genres_all'].apply(lambda x: x.split(', '))
tracks.head()
```

```python
for cntr, val in enumerate(main_genres_ids):
    tracks[genre_cols[cntr]] = tracks['track_genres_all'].apply(lambda row: 1 if str(val) in row else 0)
```

```python
tracks.head()
```

```python
# We only keep the songs with yt_views greater than 1000
tracks = tracks[tracks['yt_views'].apply(lambda x: True if x > 1000 else False)]
tracks.shape
```

```python
# The features2.csv contains mathematical characterization of audio features computed by spectral decomposition
# for each track.
# Before we merge the two dataframes, delete all the non-numerical features from the tracks dataframe.
numeric_cols = ['track_id', 'album_date_created', 'yt_views', 'genre_Experimental', 'genre_Electronic', 'genre_Rock', 'genre_Instrumental', 'genre_Pop', 'genre_Folk', 'genre_Punk', 'genre_Avant-Garde', 'genre_Hip-Hop', 'genre_Noise']
tracks = tracks[numeric_cols]
tracks.head()
```

```python
features = pd.read_csv('features2.csv')
features.head()
```

```python
# Concatenating
full_tracks = pd.concat([tracks, features], axis=1, join='inner')
```

```python
full_tracks.head()
```

```python
# We want to predict songs popularity (~youtube views) so our test data is the newest 20% of the full_tracks,
# the other 80% is split to train and validation data. The validaion data size is 20% of the full_tracks.
# Sort the values as incrementing order
full_tracks = full_tracks.sort_values('album_date_created')
full_tracks.head()
```

```python
# Selecting the subset of columns. 

full_tracks = full_tracks.filter(regex='kurtosis|genre|yt_views', axis=1)

# Add classification attribute. A song is popular if its views are beyond 10 million.
full_tracks['popular'] = full_tracks.apply(lambda row: 1 if row['yt_views'] > 200000 else 0, axis=1)

# Cut the dataset into two equal parts of two classes; beyond and below 10 million views.
full_tracks_pop = full_tracks[full_tracks['yt_views'] > 10000000]
full_tracks_notpop = full_tracks[full_tracks['yt_views'] <= 10000000]

# Create two 5k-5k datasets.
full_tracks_pop = full_tracks_pop.head(5000)
full_tracks_notpop = full_tracks_notpop.head(5000)


#display(full_tracks.head())

#display(len(full_tracks.columns))
```

```python
# Recursive feature elimination

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression, Lasso

array = full_tracks.values
X_rfe = array[:, 0:85]
Y_rfe = array[:, 85]

model_rfe = Lasso()

rfe = RFE(model_rfe, 10)
fit_rfe = rfe.fit(X_rfe, Y_rfe)
print("Num Features: ", fit_rfe.n_features_)
print("Selected Features: ", fit_rfe.support_)
print("Feature Ranking: ", fit_rfe.ranking_)

full_tracks_ranked = full_tracks.drop('popular', 1)
full_tracks_selected_columns = full_tracks_ranked.columns[fit_rfe.support_]
full_tracks_ranked = full_tracks_ranked[full_tracks_selected_columns]
full_tracks_ranked['popular'] = full_tracks['popular']
display(full_tracks_ranked.head())
```

```python
# X is the input and Y is the output
# Now drop the album_date_created column because we have previously sorted our values
#full_tracks = full_tracks.drop(columns='album_date_created')
#full_tracks = full_tracks.reset_index()
X = full_tracks_ranked.drop(columns='popular').drop(columns='yt_views')
display(X.head())
Y = full_tracks_ranked['popular']
```

```python
# Split datasets to multiple parts for training, validation and test purposes. 
nb_samples = len(X)
valid_split = 0.2
test_split = 0.2
X_train = X[0:int(nb_samples*(1-valid_split-test_split))]
Y_train = Y[0:int(nb_samples*(1-valid_split-test_split))]
X_valid = X[int(nb_samples*(1-valid_split-test_split)):int(nb_samples*(1-test_split))]
Y_valid = Y[int(nb_samples*(1-valid_split-test_split)):int(nb_samples*(1-test_split))]
X_test  = X[int(nb_samples*(1-test_split)):]
Y_test  = Y[int(nb_samples*(1-test_split)):]

X_train.head()
```

```python
# Examine the size of the new arrays.
print('len X_train: ', len(X_train))
print('len Y_train: ', len(Y_train))
print('len X_valid: ', len(X_valid))
print('len Y_valid: ', len(Y_valid))
print('len X_test: ', len(X_test))
print('len Y_test: ', len(Y_test))
```

```python
from sklearn.preprocessing import StandardScaler

# Use standard scaler to standardize features by removing the mean and scaling to unit variance.

scaler = StandardScaler().fit(X_train)

X_train = scaler.transform(X_train)
X_valid = scaler.transform(X_valid)
X_test  = scaler.transform(X_test)

# Needed to reshape the 1D arrays to 2D arrays.

Y_train = Y_train.values.reshape(-1, 1)
Y_valid = Y_valid.values.reshape(-1, 1)
Y_test = Y_test.values.reshape(-1, 1)

#scaler = StandardScaler().fit(Y_train)

#Y_train = scaler.transform(Y_train)
#Y_valid = scaler.transform(Y_valid)
#Y_test  = scaler.transform(Y_test)

display(X_train)
display(Y_train)
```

```python
from keras.callbacks import Callback
# Implementation of early stopping - Based on practice 4
# https://github.com/BME-SmartLab-Education/vitmav45/blob/master/04/XOR-Keras-callbacks.ipynb
class TrainingHistory(Callback):
    # Empty list to store metrics.
    def on_train_begin(self, logs = {}):
        # Error in training data.
        self.losses = []
        # Error in validation data.
        self.valid_losses = []
        # Metric for model accuracy on training data. 
        self.accs = []
        # # Metric for model accuracy on valdiation data. 
        self.valid_accs = []
        # Number of training epoch.
        self.epoch = 0
    
    # Saving metrics after each epoch. 
    def on_epoch_end(self, epoch, logs={}):
        self.losses.append(logs.get('loss'))
        self.valid_losses.append(logs.get('val_loss'))
        self.accs.append(logs.get('acc'))
        self.valid_accs.append(logs.get('val_acc'))
        self.epoch += 1
            
history = TrainingHistory()
```

```python
# Implemention of early stopping.

from keras.callbacks import EarlyStopping
patience = 10
early_stopping = EarlyStopping(patience = patience, verbose = 1)

# Save the best model.
from keras.callbacks import ModelCheckpoint
checkpointer = ModelCheckpoint(filepath = 'weights.hdf5', save_best_only = True, verbose = 1)

# Reduce learning rate.
from keras.callbacks import ReduceLROnPlateau
reduce_lr = ReduceLROnPlateau(monitor = 'val_loss', factor = 0.8, patience = 20, min_lr = 10e-5)
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Dropout
from keras.optimizers import SGD, Adam
from keras.callbacks import EarlyStopping

# Set random seeds.
np.random.seed(1)
from tensorflow import set_random_seed
set_random_seed(2)

from keras import regularizers

col_num = len(full_tracks_ranked.columns) - 2

def song_model(x_train, y_train, x_val, y_val, params):
    model = Sequential()

    model.add(Dense(params['first_neuron'], input_shape=(col_num,), activation=params['activation']))
    model.add(Dense(100, activation='relu'))
    #model.add(Dropout(params['dropout']))
    #model.add(Dense(1))
    model.add(Dense(1, activation=params['last_activation']))

    # We also tried different loss functions and optimizers in order to achieve better performance.
    model.compile(loss='binary_crossentropy', optimizer = params['optimizer'], metrics=['accuracy'])
    callbacks = [EarlyStopping(monitor='val_acc', patience=3, verbose=0)]
    # Fit the model
    out = model.fit(x_train, y_train, epochs = params['epochs'], validation_data=(x_val, y_val), batch_size=params['batch_size'],
                   callbacks = callbacks, verbose = 2)
    
    return out, model
```

```python
# Hyperparameter optimization
import talos
from keras.activations import relu, elu, softmax

p = {
    'first_neuron': [50, 100],
    'activation': [relu, elu],
    'last_activation': ['relu', 'sigmoid'],
    'optimizer': ['adam'],
    'batch_size': [16, 32],
    'epochs': [10],
    'dropout': [0]
    }

t = talos.Scan(X_train, Y_train,
               x_val = X_valid,
               y_val = Y_valid,
               params = p,
               model = song_model,
               grid_downsample = .1)
```

```python
t.data.head()

t.saved_models

r = talos.Reporting(t)

r.best_params()
```
